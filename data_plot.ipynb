```python
import matplotlib.pyplot as plt   #导入模块matplotlib.pyplot，并重新命名为plt

squares = [1,4,9,16,25,36,49,64,81,100]  #定义一个数组

plt.plot(squares, linewidth=5)# 设置图表标题，并给坐标轴加上标签和 参数 linewidth 决定了绘制的线条的粗细
plt.title("Square Numbers", fontsize=24)#设置标题和字体大小
plt.xlabel("Value", fontsize=14)  #  x轴标签，和字体大小
plt.ylabel("Square of Value", fontsize=14)  #  y轴标签，和字体大小
plt.tick_params(axis='both', labelsize=14) # 设置刻度标记的大小，函数 tick_params() 设置刻度的样式
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('result/log (1).txt',encoding='utf-8',sep='\t')
df.columns = ['Train Loss','Train Loss X','Train Loss U','Train Loss S','Test Loss','Test Acc','NAN']
del df['NAN']
df
```

```python
import copy

df1 = copy.deepcopy(df).round(2)
df1
```

```python
plt.figure(figsize=(12,5),dpi=200)
loss = df1.drop('Test Acc',axis=1)
loss['index'] = range(1,len(loss)+1)

x_ticks = np.arange(0, len(loss), 1)

plt.plot(loss['Train Loss'],label='Train Loss')
plt.plot(loss['Train Loss X'],label='Train Loss X')
plt.plot(loss['Train Loss U'],label='Train Loss U')
plt.plot(loss['Train Loss S'],label='Train Loss S')
plt.plot(loss['Test Loss'],label='Test Loss')

plt.xlabel('Epoch',fontsize=14)
plt.ylabel('Loss',fontsize=14)
plt.xticks(x_ticks)

for a,b in zip(x_ticks,loss['Train Loss']):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

for a,b in zip(x_ticks,loss['Test Loss']):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

plt.legend()
plt.savefig('data_analysis_pic/loss.png')
plt.show()
```

```python
plt.figure(figsize=(12,5),dpi=200)
acc = df1.drop(['Train Loss','Train Loss X','Train Loss U','Train Loss S','Test Loss'],axis=1)
acc['index'] = range(1,len(acc)+1)
plt.plot(acc['Test Acc'],label='Test Acc',color='red',marker='o',markersize=5)
plt.xlabel('Epoch',fontsize=14)
plt.ylabel('Acc',fontsize=14)
plt.xticks(x_ticks)
plt.title('Test Acc: un_label:4000, label: 8271',fontsize=14,pad=20)

for a,b in zip(x_ticks,acc['Test Acc']):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10,rotation=45,
             bbox=dict(facecolor='red', alpha=0.05))

plt.legend(loc='lower right')
plt.savefig('data_analysis_pic/acc.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5),dpi=200)

loss = df1.drop('Test Acc',axis=1)
loss['index'] = range(1,len(loss)+1)

x_ticks = np.arange(1, len(loss)+1, 1)
print(x_ticks)

ax.plot(x_ticks,loss['Train Loss'],label='Train Loss',color='slategrey',marker='^',markersize=5)
ax.plot(x_ticks,loss['Train Loss X'],label='Train Loss X',color='cornflowerblue',markersize=5)
ax.plot(x_ticks,loss['Train Loss U'],label='Train Loss U',color='darkorange',marker='s',markersize=5)
ax.plot(x_ticks,loss['Train Loss S'],label='Train Loss S',color='purple',marker='*',markersize=5)
ax.plot(x_ticks,loss['Test Loss'],label='Test Loss',color='darkgreen',marker='o',markersize=5)

ax.set_xlabel('Epoch',fontsize=14)
ax.set_ylabel('Loss',fontsize=14)
ax.set_xticks(x_ticks)

for a,b in zip(x_ticks,loss['Train Loss']):
    ax.text(a, b, b, ha='center', va='bottom', fontsize=10)

for a,b in zip(x_ticks,loss['Test Loss']):
    ax.text(a, b, b, ha='center', va='bottom', fontsize=10)

ax.legend(loc=(0.2,0.5))
ax.spines['right'].set_visible(False) # ax右轴隐藏



z_ax = ax.twinx() # 创建与轴群ax共享x轴的轴群z_ax

acc = df1.drop(['Train Loss','Train Loss X','Train Loss U','Train Loss S','Test Loss'],axis=1)
acc['index'] = range(1,len(acc)+1)
z_ax.plot(x_ticks, acc['Test Acc'],label='Test Acc',color='red',marker='o',markersize=5)
# z_ax.set_xlabel(x_ticks, 'Epoch',fontsize=14)
z_ax.set_ylabel('Acc',fontsize=14)
z_ax.set_xticks(x_ticks)
plt.title('Test Acc: un_label:4000, label: 8271',fontsize=14,pad=20)

for a,b in zip(x_ticks,acc['Test Acc']):
    z_ax.text(a, b, b, ha='center', va='bottom', fontsize=10,rotation=45,
             bbox=dict(facecolor='red', alpha=0.02))

z_ax.legend(loc=(0.85,0.8))




plt.savefig('data_analysis_pic/loss_acc.png')

plt.show()
```
